## Imports

```python
import urllib3 as urllib
from bs4 import BeautifulSoup
import requests
import urllib.request
import pandas as pd
```

## Scraping

```python
# function to scrape twitter urls from https://covidtracking.com/data/


def scrape_urls(url):
    # get html
    req = urllib3.PoolManager()
    res = req.request('GET', url)
    soup = BeautifulSoup(res.data, 'html.parser')
    accounts = soup.find_all('div', class_= 'state-header')
    
    # empty list to append [state, twitter_url]
    l = []
    
    for account in accounts:
        # get state
        state_name = account.find('h2').find("a").text
        
        # get twitter url
        twitter_url = account.find_all("a", {'class', 'state-link'})[1].get('href')
        
        # append to list
        l.append([state_name, twitter_url])
    
    # create df and drop American Samoa (no twitter)
    df = pd.DataFrame(columns=['state', 'twitter_url'], data = l)
    df = df.drop(df.index[3])

    return df

# specify the url
url = "https://covidtracking.com/data/"

# scrape and generate csv
df = scrape_urls(url)
df.to_csv('twitter_accts.csv', index=False)

df
```
